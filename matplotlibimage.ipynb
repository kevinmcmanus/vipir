```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import vipir as vp
import matplotlib
from matplotlib import cm
import numpy as np
```

```python
from matplotlib.patches import Rectangle
```

```python
cdf = vp.vipir(r"./netcdf/WI937_2020004132603.NGI")
```

```python
snr = cdf.total_power-cdf.total_noise.reshape(-1,1)

fig, ax = plt.subplots(figsize=(16,16))
pcm = ax.pcolormesh( cdf.freq, cdf.rng, snr.T, vmin=0, vmax=100,
                    cmap='gnuplot')

r = Rectangle((2000,200),3300, 500, edgecolor='r', lw=5, facecolor='none')
ax.add_patch(r)
ax.set_ylabel('Range')
ax.set_xlabel('Frequency')
ax.set_xscale('log')
ax.set_xlim(1000,25000)
cbar=fig.colorbar(pcm, ax=ax)

cbar.set_label('Signal to Noise Ratio (dB)', rotation=270, labelpad=30)
ax.set_title('Station: {0}'.format(cdf.station), fontsize=16)
```
